```python
2: An example that prints the python exception hierarchy using
   'logging' module:

import logging

# Configure logging
logging.basicConfig(filename='example.log', level=logging.DEBUG)

try:
    num1 = int(input("Enter a number: "))
    num2 = int(input("Enter another number: "))
    result = num1 / num2
    print("The result is:", result)
except ValueError:
    logging.error("Invalid input")
except ZeroDivisionError:
    logging.error("Cannot divide by zero")

>first import the logging module to enable logging.
>then,configure logging to write to a file named example.log and 
 set the logging level to DEBUG.
> I use a try block to enclose the code that may raise exceptions.
> I prompt the user to enter two integers num1 and num2.
> then i  attempt to perform division operation on num1 and num2 and store
  the result in result.
> print the result if the division was successful.
> the i use except statements to catch specific exceptions that may occur in
  the try block.
> If a 'ValueError' occurs, we log an error message indicating that the
  input was invalid.
> if  'ZeroDivisionError' occurs, we log an error message indicating that we cannot divide by zero.

      
    
    
```

```python
3: The 'ArithmeticError' class is a built-in exception class in Python that
   is raised for arithmetic errors, such as division by zero or overflow.
    
Here are two common errors defined in the ArithmeticError class

1."ZeroDivisionError"
This error is raised when you attempt to divide by zero.
Example:
 
a = 10
b = 0
try:
    result = a / b
except ZeroDivisionError as e:
    print(" ZeroDivisionError:", e)
    

In this example, I have define two variables a and b, where b is zero.
When we attempt to divide a by b, a ZeroDivisionError is raised. 
We catch the error in the except block and print the error message.


2. "OverflowError"

This error is raised when a calculation exceeds the maximum representable
value for a numeric type.

Example:

import sys

try:
    result = sys.maxsize * 2
    print(result)
except OverflowError:
    print("OverflowError: Result too large to represent as an integer")

    
In this example, Iam trying to calculate the result of multiplying the
maximum value representable as an integer by 2. This calculation results in
a number that is too large to represent as an integer, which raises an 
"OverflowError". We catch this error using a try-except block and print a
message to indicate the error.



```

```python
4: The LookupError class is used as the base class for exception raised when
an index or key used to access a sequence or dictionary is invalid.
 It is a subclass of the Exception class and is further subclassed into
specific lookup-related exceptions such as IndexError and KeyError.

# IndexError example
my_list = [1, 2, 3]
try:
    value = my_list[3]  # index 3 is out of range
except IndexError:
    print("Error: Index out of range")

In this example, we have a list with three elements, but i try to access
the element at index 3, which is out of range. This raises an "IndexError", 
which i catch using a try-except block and print an error message.


# KeyError example
my_dict = {'a': 1, 'b': 2, 'c': 3}
try:
    value = my_dict['d']  # key 'd' does not exist
except KeyError:
    print("Error: Key not found")


In this example, we have a dictionary with keys 'a', 'b', and 'c'. i have
try to access the value associated with the key 'd', which does not exist in
the dictionary. This raises a "KeyError", which we catch using a try-except
block and print an error message.

```

```python
5:"import Error"  is a built-in Python exception that is raised when a 
module or package cannot be imported for any reason.

This can happen if the module or package does not exist, or if there is an
error in the code within the module or package. ImportError is a catch-all
exception that can be used to handle import-related errors in a generalized
way. By catching and handling 'ImportError' exceptions, you can provide more
meaningful error messages to users or take appropriate actions to resolve the error.

"ModuleNotFoundError"
it is a subclass of the ImportError exception that is raised when a module
or package cannot be found by the Python interpreter. This usually occurs 
when you try to import a module that doesnt exist or has not been installed
yet. The ModuleNotFoundError exception was introduced in Python 3.6 as a
more specific error message for missing modules, making it easier to 
diagnose and fix import-related issues. When catching this exception, you 
can provide a more targeted error message to users or take appropriate
actions to resolve the error.


```
